This notebook explains how to enhance plots with titles, axis labels, legends, and annotations. These elements make visualizations clearer, more descriptive, and easier to interpret.

## 1. Why Titles, Labels & Annotations Matter

A plot without titles or labels is incomplete. These elements help:
- Describe what the graph represents
- Identify the meaning of x-axis and y-axis
- Highlight key points or events in the graph
- Provide context for readers

Matplotlib provides simple functions to add all of these elements.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Adding a Title

The function `plt.title()` or `ax.set_title()` is used to add a title.

**Common Parameters:**
- `fontsize`: size of the title
- `color`: title color
- `loc`: left, right, or center
- `pad`: spacing between title and plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(7,4))
plt.plot(x, y)
plt.title("Sine Wave Plot", fontsize=14, pad=10)
plt.show()
```

## 3. Adding Axis Labels

Axis labels describe what each axis represents.

**Functions:**
- `plt.xlabel()` → Label for x-axis
- `plt.ylabel()` → Label for y-axis

```python
plt.figure(figsize=(7,4))
plt.plot(x, y)
plt.title("Sine Function")
plt.xlabel("X Values (radians)", fontsize=12)
plt.ylabel("sin(x)", fontsize=12)
plt.show()
```

## 4. Adding Legends

Legends explain what each line or marker represents.

**Two steps:**
1. Add a `label` inside `plt.plot()`
2. Call `plt.legend()`

```python
plt.figure(figsize=(7,4))
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.title("Sine and Cosine Functions")
plt.xlabel("X Values")
plt.ylabel("Function Output")
plt.legend()
plt.show()
```

## 5. Adding Annotations

Annotations highlight important points using:
- arrows
- text descriptions
- markers

**Common function:**
`plt.annotate(text, xy, xytext, arrowprops={})`

```python
plt.figure(figsize=(7,4))
plt.plot(x, y)

# Find max value of sin(x)
x_max = np.pi/2
y_max = np.sin(x_max)

plt.scatter(x_max, y_max, color="red")
plt.annotate(
    "Maximum Point",
    xy=(x_max, y_max),
    xytext=(2, 1),
    arrowprops={"arrowstyle": "->"}
)

plt.title("Sine Function with Annotation")
plt.xlabel("X Values")
plt.ylabel("sin(x)")
plt.show()
```

## 6. Adding Multiple Texts on Plot

You can place text anywhere using `plt.text()`.

```python
plt.figure(figsize=(7,4))
plt.plot(x, y)
plt.text(5, 0.5, "Middle of Plot", fontsize=12)
plt.title("Adding Text on Plot")
plt.show()
```

## 7. Summary
In this notebook, you learned how to enhance Matplotlib visualizations using:
- Titles (`plt.title`)
- Axis Labels (`plt.xlabel`, `plt.ylabel`)
- Legends (`plt.legend`)
- Annotations (`plt.annotate`)
- Custom text (`plt.text`)

These tools significantly improve the readability and effectiveness of your plots.
