Matplotlib is structured around two fundamental objects:
- **Figure:** The entire drawing canvas (a full page or window).
- **Axes:** The actual plot area where charts are drawn.

Understanding these two components is essential for mastering Matplotlib.

## What is a Figure?
A `Figure` is the top-level container for all plot elements. It may contain:
- One or multiple Axes
- Titles
- Legends
- Text elements
- Images and other artists

Think of a `Figure` as a **blank page** on which one or more plots (Axes) are placed.

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6,4))
plt.show()
```

## What is Axes?
`Axes` represents the main area where data is visualized. It contains:
- x-axis & y-axis
- Tick labels
- Plot elements like lines, bars, scatter points, etc.

Think of `Axes` as the **graph itself** inside the page.

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot([1,2,3], [4,5,6])
ax.set_title("Simple Line Plot")
plt.show()
```

## Figure vs Axes Comparison
| Component | Description |
|----------|-------------|
| **Figure** | The whole drawing canvas |
| **Axes** | The actual area where the plot appears |
| **Axis** | The x and y axis within the Axes |
| **Artist** | Anything visible (lines, text, shapes) |

## Multiple Axes Example
A single Figure can host multiple subplots.

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))
axs[0].plot([1,2,3], [3,2,5])
axs[1].bar([1,2,3], [5,3,4])
axs[0].set_title("Line Plot")
axs[1].set_title("Bar Plot")
plt.show()
```

## Multiple Figures Example
You can create multiple independent figures in a script.

```python
fig1 = plt.figure(figsize=(5,3))
plt.plot([1,2,3],[3,2,1])
plt.title("Figure 1")
plt.show()

fig2 = plt.figure(figsize=(5,3))
plt.plot([1,2,3],[1,4,9])
plt.title("Figure 2")
plt.show()
```

## Summary
- A **Figure** is the overall container.
- **Axes** is where plots are drawn.
- Multiple Axes can exist inside one Figure.
- Multiple Figures can be created in a script.
