## ----------------------------------------------------------
## Importing Matplotlib
## ----------------------------------------------------------
The most common way to import Matplotlib is through the **pyplot** interface.

````python
import matplotlib.pyplot as plt
import matplotlib

print("Matplotlib version:", matplotlib.__version__)
````

## üìå First Simple Plot
Let's create our first basic line plot using Matplotlib.

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
````

## üìö Understanding Pyplot
`pyplot` works like a **state machine**, meaning:
- When you call `plt.plot()` ‚Üí Matplotlib automatically creates a figure.  
- Calling `plt.title()` applies to the current active figure.

It is designed to work simply and quickly.
