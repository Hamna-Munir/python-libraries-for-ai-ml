## 1. Adding Titles
A **title** describes the overall purpose of the plot.
You can add a title in two ways:

### Using Pyplot
```python
plt.title("My Plot Title")
```
### Using Object-Oriented API
```python
fig, ax = plt.subplots()
ax.set_title("My Plot Title")
```
**Important Parameters:**
- `fontsize`: Adjusts font size
- `fontweight`: Makes title bold
- `loc`: `'left'`, `'center'`, `'right'`
- `pad`: Space between title and plot


````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1,2,3], [2,4,3])
ax.set_title("Basic Line Plot Title", fontsize=14, fontweight='bold', loc='center')
plt.show()
````

## 2. Adding Axis Labels
**Axis labels** explain what each axis represents.

**Pyplot Version:**
```python
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")
```
**OOP Version:**
```python
ax.set_xlabel("X-Axis Label")
ax.set_ylabel("Y-Axis Label")
```
**Useful Parameters:**
- `fontsize`
- `labelpad`: Space between axis and label


````python
fig, ax = plt.subplots()
ax.plot([0,1,2,3], [5,6,2,7])
ax.set_xlabel("Time (seconds)", fontsize=12, labelpad=10)
ax.set_ylabel("Value", fontsize=12, labelpad=10)
plt.show()
````

## 3. Adding Text Inside Plots (`text`)
Use `text()` to place custom text anywhere.

```python
ax.text(x, y, "Some Text", fontsize=12, color="red")
```
**Common Uses:**
- Labeling data points
- Adding comments
- Writing values directly on the chart


````python
fig, ax = plt.subplots()
ax.plot([1,2,3], [3,1,4])
ax.text(2, 1, "Low Point", fontsize=12)
plt.show()
````

## 4. Adding Annotations (`annotate`)
`annotate()` is more powerful than `text()` because it allows **arrows** and **pointing**.

### Syntax
```python
ax.annotate("Label", xy=(x, y), xytext=(x2, y2), arrowprops={})
```
**Important Parameters:**
- `xy`: Point to highlight
- `xytext`: Text location
- `arrowprops`: Style of arrow
    - `arrowstyle`
    - `color`
    - `lw`: Line width


````python
fig, ax = plt.subplots()
x = [1,2,3,4]
y = [2,3,6,4]
ax.plot(x, y)

ax.annotate(
    "Peak Value",
    xy=(3, 6),
    xytext=(2.5, 7),
    arrowprops={"arrowstyle": "->", "lw": 1.5}
)
plt.show()
````

## 5. Title, Labels, and Annotation Combined Example
This example shows **all customization features together**.

````python
fig, ax = plt.subplots(figsize=(6,4))

x = [1,2,3,4,5]
y = [3,5,2,8,7]
ax.plot(x, y, marker='o')

# Title
ax.set_title("Sales Trend Analysis", fontsize=15, fontweight='bold')

# Axis Labels
ax.set_xlabel("Quarter")
ax.set_ylabel("Sales (in thousands)")

# Annotation for highest point
ax.annotate(
    "Peak Sales",
    xy=(4, 8),
    xytext=(3.2, 9),
    arrowprops={"arrowstyle": "->", "lw": 1.4}
)

plt.show()
````
