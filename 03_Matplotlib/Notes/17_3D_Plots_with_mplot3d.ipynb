## Introduction
Matplotlib supports 3D plotting through the **`mpl_toolkits.mplot3d`** module. 3D visualizations help:
- Understand complex surfaces and spatial data
- Visualize scientific simulations
- Explore machine learning data in three dimensions

In this notebook, we will cover:
- 3D line plots
- 3D scatter plots
- 3D surface plots
- Customizing 3D plots (colors, markers, labels, and viewpoints)

# 1. Setting up 3D Axes
To create a 3D plot, you need to import **`Axes3D`** and create a 3D axes object.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_title('3D Axes Setup')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

# 2. 3D Line Plot
3D line plots show trajectories or trends in three dimensions.

```python
# Sample data
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='3D Line', color='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Line Plot')
ax.legend()
plt.show()
```

# 3. 3D Scatter Plot
Scatter plots are useful for visualizing points in 3D space.

```python
# Sample 3D data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
colors = np.random.rand(50)
sizes = 50 * np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=colors, s=sizes, cmap='viridis', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')
fig.colorbar(scatter, ax=ax, label='Color Intensity')
plt.show()
```

# 4. 3D Surface Plot
Surface plots visualize 3D functions or grids.

```python
# Create grid
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='k', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5, label='Z Value')
plt.show()
```

# 5. Customizing View Angle
You can rotate the 3D plot to view from different angles using `view_init`.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.view_init(elev=30, azim=60)  # Adjust elevation and azimuth
ax.set_title('3D Surface Plot – Custom View')
plt.show()
```

# Summary
- **3D Axes** are created using `projection='3d'`.
- **Line plots** visualize trajectories in space.
- **Scatter plots** display points with optional colors and sizes.
- **Surface plots** represent 3D functions or grids.
- **Customization**: colors, labels, grid, legends, and view angles.

3D plotting in Matplotlib is useful for spatial data visualization, scientific computing, and exploratory data analysis.
