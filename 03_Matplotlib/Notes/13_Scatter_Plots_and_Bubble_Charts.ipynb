## Introduction
Scatter plots are used to visualize relationships between two numerical variables. They are essential for:
- Identifying patterns or trends
- Detecting correlations
- Highlighting clusters or outliers

Bubble charts are an extension of scatter plots where each point also has a *size*, helping you represent a third dimension.

In this notebook, we cover:
- Creating basic scatter plots
- Customizing marker shapes, colors, and sizes
- Adding transparency
- Creating bubble charts with variable sizes
- Using colormaps inside scatter plots
- Adding annotations to scatter visualizations

This is one of the most powerful visualization tools for data analysis, especially for understanding relationships and distributions.

# 1. Creating a Basic Scatter Plot
A scatter plot requires two numerical arrays:
- **x-values** → horizontal axis
- **y-values** → vertical axis

Matplotlib function:
```python
plt.scatter(x, y)
```

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title("Basic Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
````

# 2. Customizing Marker Styles in Scatter Plots
Scatter plot markers can be customized using:
- `marker='o'` → marker type
- `s=100` → size
- `color='red'` → color
- `edgecolors='black'` → border color

````python
plt.scatter(x, y, marker='o', s=120, color='orange', edgecolors='black')
plt.title("Customized Scatter Plot")
plt.show()
````

# 3. Adding Transparency (alpha)
`alpha` helps reduce overlapping effects.
- `alpha=1.0` → fully opaque
- `alpha=0.3` → transparent


````python
plt.scatter(x, y, s=150, alpha=0.5, color='purple')
plt.title("Scatter Plot with Transparency")
plt.show()
````

# 4. Creating a Bubble Chart (Using Size as Third Variable)
Bubble charts show a **third dimension** using marker size:

```python
plt.scatter(x, y, s=size_variable)
```

````python
sizes = np.random.randint(50, 500, size=50)  # bubble sizes

plt.scatter(x, y, s=sizes, alpha=0.6, color='skyblue', edgecolors='black')
plt.title("Bubble Chart Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
````

# 5. Using Colormaps in Scatter Plots
You can use data values to control color intensity using:
```python
c=value_array
cmap='viridis'
```

````python
colors = np.random.rand(50)

plt.scatter(x, y, c=colors, cmap='viridis', s=200)
plt.colorbar(label='Color Scale')
plt.title("Scatter Plot with Colormap")
plt.show()
````

# 6. Combining Color + Size for 4-Dimensional Visualization
You can visualize:
- x (horizontal)
- y (vertical)
- size (bubble size)
- color (value intensity)


````python
sizes = np.random.randint(100, 600, size=50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, cmap='plasma', alpha=0.7, edgecolors='black')
plt.colorbar(label='Color Values')
plt.title("4D Bubble Chart (X, Y, Size, Color)")
plt.show()
````

# 7. Adding Labels / Annotations to Scatter Points
Annotations help identify specific points in scatter plots.

Useful function:
```python
plt.annotate(text, (x, y))
```

````python
plt.scatter(x, y, s=180, color='teal')
plt.annotate("Important Point", (x[0], y[0]), textcoords="offset points", xytext=(10,10))
plt.title("Scatter Plot with Annotations")
plt.show()
````

# Summary
- Scatter plots show relationships between two numerical variables.
- Marker styles, sizes, edge colors, and transparency increase readability.
- Bubble charts add a *third dimension* using marker size.
- Colormaps introduce color gradients to represent additional data.
- You can visualize 4D data using x, y, size, and color.
- Annotations help highlight important points.

Scatter and bubble charts are powerful tools for visualizing patterns, correlations, and multi-dimensional data.
