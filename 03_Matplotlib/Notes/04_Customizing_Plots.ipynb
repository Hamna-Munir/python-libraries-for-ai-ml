## 1. Introduction to Plot Customization
Matplotlib allows full control over plot appearance including:
- Titles
- Axis labels
- Line styles and widths
- Colors
- Markers
- Legends
- Grid lines
- Figure size

Understanding customization helps you create visualizations that communicate information more effectively.


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

---
## 2. Changing Figure Size
You can adjust the figure size using `figsize`. This is useful for reports, dashboards, and presentations.

`plt.figure(figsize=(width, height))`

```python
plt.figure(figsize=(8, 4))
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Custom Figure Size Example")
plt.show()
```

---
## 3. Customizing Line Style, Width, and Color
Matplotlib provides several customization options for lines:

- `color` → sets line color
- `linewidth` or `lw` → controls thickness
- `linestyle` or `ls` → controls line style

**Common Line Styles:**
- '-'  (solid)
- '--' (dashed)
- '-.' (dash-dot)
- ':'  (dotted)


```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.figure(figsize=(7, 4))
plt.plot(x, y, color='green', linewidth=2, linestyle='--')
plt.title("Custom Line Style Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

---
## 4. Adding Markers
Markers highlight individual data points.

**Common Markers:**
- 'o' (circle)
- 's' (square)
- '^' (triangle up)
- 'x' (cross)
- '*' (star)

You can combine markers with lines for better visibility.

```python
plt.figure(figsize=(7, 4))
plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.title("Line Plot with Markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

---
## 5. Adding Titles and Axis Labels
Titles help describe the plot, while axis labels explain the meaning of each axis.

Functions used:
- `plt.title()`
- `plt.xlabel()`
- `plt.ylabel()`

```python
plt.figure(figsize=(7, 4))
plt.plot(x, y)
plt.title("Quadratic Growth Graph")
plt.xlabel("Input Value")
plt.ylabel("Squared Output")
plt.show()
```

---
## 6. Adding Grid Lines
Grid lines help in reading values accurately.

`plt.grid(True)` enables the grid.
`plt.grid(axis='x')` enables only x-axis grid.


```python
plt.figure(figsize=(7, 4))
plt.plot(x, y)
plt.title("Plot with Grid Lines")
plt.grid(True)
plt.show()
```

---
## 7. Adding Legends
Legends help identify different lines or categories in a plot.

To use legends:
- Add `label=` inside `plt.plot()`
- Use `plt.legend()` to display it


```python
plt.figure(figsize=(7, 4))
plt.plot(x, y, label="y = x^2", color='purple')
plt.plot(x, x, label="y = x", color='orange')
plt.title("Legend Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

---
## 8. Changing Axis Limits
Axis limits help to zoom in or out of the plot.

Functions:
- `plt.xlim(min, max)`
- `plt.ylim(min, max)`

```python
plt.figure(figsize=(7, 4))
plt.plot(x, y)
plt.title("Custom Axis Limits")
plt.xlim(0, 5)
plt.ylim(0, 20)
plt.grid(True)
plt.show()
```
