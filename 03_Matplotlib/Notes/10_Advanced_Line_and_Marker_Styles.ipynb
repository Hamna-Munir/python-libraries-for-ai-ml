# 1. Advanced Line Styles
Matplotlib supports several built-in line style patterns:
- `'-'` → Solid line
- `'--'` → Dashed line
- `'-.'` → Dash-dot line
- `':'` → Dotted line

You can also create custom dash patterns using `set_dashes()`.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)

plt.plot(x, np.sin(x), linestyle='-', label='Solid')
plt.plot(x, np.cos(x), linestyle='--', label='Dashed')
plt.plot(x, np.sin(x*0.5), linestyle='-.', label='Dash-dot')
plt.plot(x, np.cos(x*0.5), linestyle=':', label='Dotted')

plt.title("Built-in Line Styles")
plt.legend()
plt.show()
```

# 2. Custom Dash Patterns
Custom dash sequences allow complete control over line appearance.

**Pattern Format:** `[dash_length, gap_length, dash_length, gap_length, ...]`

```python
line, = plt.plot(x, np.sin(x), label='Custom Dash')
line.set_dashes([10, 5, 2, 5])  # long dash, short gap, short dash, gap

plt.title("Custom Dash Pattern")
plt.legend()
plt.show()
```

# 3. Marker Styles
Markers indicate individual data points and are extremely useful in scatter-type visualizations.

Common marker styles include:
- `'o'` → Circle
- `'s'` → Square
- `'^'` → Triangle
- `'+'` → Plus
- `'*'` → Star
- `'x'` → X marker

Markers can be combined with lines or used alone in scatter plots.

```python
plt.plot(x, np.sin(x), marker='o', markersize=6, label='Circle')
plt.plot(x, np.cos(x), marker='s', markersize=6, label='Square')
plt.plot(x, np.sin(x*0.3), marker='^', markersize=6, label='Triangle')

plt.title("Different Marker Styles")
plt.legend()
plt.show()
```

# 4. Customizing Marker Size, Colors, and Edges
Markers offer full styling options:
- `markersize` → marker size
- `markerfacecolor` → fill color
- `markeredgecolor` → border color
- `markeredgewidth` → border thickness

```python
plt.plot(x, np.sin(x), marker='o', markersize=10,
         markerfacecolor='white', markeredgecolor='blue', markeredgewidth=2,
         label='Customized Marker')

plt.title("Marker Customization")
plt.legend()
plt.show()
```

# 5. Combining Line Styles and Marker Styles
You can mix line and marker customizations together for better clarity.

```python
plt.plot(x, np.sin(x), linestyle='--', linewidth=2,
         marker='o', markersize=8, markerfacecolor='yellow',
         label='Line + Marker')

plt.title("Combining Line and Marker Styles")
plt.legend()
plt.show()
```

# 6. Line Width, Cap Styles, and Join Styles
Advanced styling options for professional-quality graphics:

### **Line Width (`linewidth`)**
Controls thickness.

### **Cap Styles (`solid_capstyle`)**
- `'butt'`
- `'round'`
- `'projecting'`

### **Join Styles (`solid_joinstyle`)**
- `'miter'`
- `'round'`
- `'bevel'`

```python
plt.plot(x, np.sin(x), linewidth=4,
         solid_capstyle='round', solid_joinstyle='bevel',
         label='Styled Line')

plt.title("Line Width, Cap Style, Join Style")
plt.legend()
plt.show()
```

# 7. Styling Multiple Lines Automatically
Matplotlib allows automatic cycling of line styles using style cycles. This is helpful when plotting many lines together.

```python
from itertools import cycle

line_styles = cycle(['-', '--', '-.', ':'])

for i in range(4):
    plt.plot(x, np.sin(x + i), linestyle=next(line_styles), label=f'Line {i+1}')

plt.title("Automatic Line Style Cycling")
plt.legend()
plt.show()
```
