## üîß 2. Importing Matplotlib
The most commonly used module of Matplotlib is **pyplot**, imported as `plt`.


````python
import matplotlib.pyplot as plt

print("Matplotlib Imported Successfully")
````

## üñ• 3. Checking Matplotlib Version
Always check the version to ensure compatibility.

````python
import matplotlib
print("Matplotlib Version:", matplotlib.__version__)
````

## üé® 4. Setting Up Jupyter Notebook for Inline Plots
Jupyter Notebook needs a magic command to show charts **inside the notebook**.

Use the following command:

````python
%matplotlib inline
````

`%matplotlib inline` ensures that:
- Plots appear directly in the notebook
- No extra window opens
- Cleaner & faster workflow

---
## üñº 5. Creating a Simple Test Plot
Let's verify that Matplotlib is working correctly.

````python
plt.figure(figsize=(6,4))
plt.plot([1,2,3,4], [2,4,6,8])
plt.title("Test Plot: Matplotlib Setup Working")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
````

## üß∞ 6. Matplotlib Backends
Matplotlib has different **backends** used to render graphics.

Two types:
- **Interactive backends** ‚Üí open separate windows (Qt, Tkinter)
- **Non-interactive backends** ‚Üí used in Jupyter Notebook (inline)

Check the current backend:

````python
print("Current backend:", matplotlib.get_backend())
````

Using `inline` backend is ideal for notebooks.
