## Introduction
Color is one of the most important elements in data visualization. Matplotlib provides:
- Built-in colors
- Custom colors using HEX/RGB
- Colormaps (for continuous or categorical data)
- Tools to adjust transparency and intensities

In this notebook, you will learn how to work with colors effectively to make plots meaningful and visually appealing.

# 1. Using Basic Colors in Matplotlib
Matplotlib supports multiple color formats:

### a. Named Colors
Examples: `"red"`, `"blue"`, `"green"`, `"yellow"`

### b. HEX Colors
Example: `"#1f77b4"`

### c. RGB Values (0–1 range)
Example: `(0.2, 0.5, 0.7)`

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), color='blue', label='Named Color')
plt.plot(x, np.cos(x), color='#FF5733', label='HEX Color')
plt.plot(x, np.tan(x)/10, color=(0.2, 0.5, 0.7), label='RGB Color')

plt.title("Using Different Color Formats")
plt.legend()
plt.show()
```

# 2. Adding Transparency with Alpha
The `alpha` parameter controls opacity.
- `alpha=1.0` → fully opaque
- `alpha=0.0` → fully transparent

```python
plt.plot(x, np.sin(x), color='red', alpha=0.3, label='alpha=0.3')
plt.plot(x, np.sin(x), color='red', alpha=1.0, label='alpha=1.0')

plt.title("Transparency with Alpha Values")
plt.legend()
plt.show()
```

# 3. Working with Colormaps
Colormaps are sequences of colors used to represent data gradients.
Matplotlib includes many built-in colormaps, divided into:

### 1. Sequential Colormaps
Used for ordered data (e.g., `viridis`, `plasma`, `inferno`).

### 2. Diverging Colormaps
Used when data has a critical midpoint (e.g., `coolwarm`, `bwr`).

### 3. Categorical Colormaps
Used for unrelated categories (e.g., `tab10`, `Set1`).

List all available colormaps:

```python
plt.colormaps()
```

# 4. Applying a Colormap to Scatter Plots
Colormaps are useful when representing continuous values.


```python
y = np.sin(x)
colors = y  # using values as color

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar(label='Value')
plt.title("Scatter Plot with Viridis Colormap")
plt.show()
```

# 5. Using Colormaps in Line Plots
You can color each segment of the line differently using colormaps.

```python
colors = plt.cm.plasma(np.linspace(0, 1, len(x)))

for i in range(len(x)-1):
    plt.plot(x[i:i+2], y[i:i+2], color=colors[i])

plt.title("Line Plot with Plasma Colormap Gradient")
plt.show()
```

# 6. Reversing Colormaps
Any colormap can be reversed by adding `_r` at the end.

Examples:
- `viridis_r`
- `inferno_r`
- `coolwarm_r`

```python
plt.scatter(x, y, c=y, cmap='coolwarm_r')
plt.colorbar()
plt.title("Reversed Coolwarm Colormap")
plt.show()
```

# 7. Creating Custom Colormaps
You can create your own colormaps using:
- A list of HEX/RGB colors
- `LinearSegmentedColormap`


```python
from matplotlib.colors import LinearSegmentedColormap

colors_list = ['#0000FF', '#00FF00', '#FF0000']  # blue → green → red
custom_cmap = LinearSegmentedColormap.from_list('custom_cmap', colors_list)

plt.scatter(x, y, c=y, cmap=custom_cmap)
plt.colorbar()
plt.title("Custom Blue-Green-Red Colormap")
plt.show()
```

# Summary
- Matplotlib provides multiple ways to use colors: names, HEX, RGB, and transparency.
- Colormaps help visualize continuous, sequential, and categorical data.
- Built-in colormaps include `viridis`, `plasma`, `coolwarm`, `tab10`, etc.
- Colormaps can be reversed using `_r`.
- Custom colormaps can be created using color lists.

Understanding color usage is essential for clear, informative visualizations.
